# Задание 1
Необходимо с помощью SQL запроса собрать одну общую таблицу из всех данных витрин, которая должна включать в себя все поля этих витрин.
В качестве ответа предоставить ссылку на вашу копию страницы, где можно увидеть написанный запрос и собранную таблицу. Ссылку вставьте в Word файл "Техническое задание" в соответствующее поле.


_(одна из таблиц называется `group`, её имя конфликтует с зарезервированными ключевым словом SQL, поэтому его нужно будет заключить в двойные кавычки `""`)_


## Решение

```python
%%sql
create table QWE AS -- сохраняем в таблицу для дальнейшего использования
WITH cte1 AS(
	SELECT *, call.finish_dttm AS comm_finish_dttm  -- одинаковое название у разных переменных -> переименовываем
	FROM task
	JOIN call
		ON task.task_rk=call.wo_task_rk
	LEFT JOIN action
		ON call.wo_hit_rk=action.hit_rk
	LEFT JOIN product
		ON call.wo_hit_rk=product.hit_rk
	LEFT JOIN emp_x_org_gr
		ON call.wo_employee_rk=emp_x_org_gr.employee_rk
	LEFT JOIN result
		ON action.hit_status_result_id=result.hit_status_result_id
	LEFT JOIN queue
		ON call.wo_queue_id=queue.queue_id
	LEFT JOIN 'group'
		ON emp_x_org_gr.org_group_rk='group'.org_group_rk
	LEFT JOIN mngmnt
		ON 'group'.org_management_rk=mngmnt.org_management_rk
	LEFT JOIN horoscope
		ON call.agent_login=horoscope.agent_login
	)
SELECT task_rk, task_stage_id, source_system_cd, create_dttm, -- т.к. функция "select * exept..." в sqlite не поддерживается, удаляем дубликаты следующим образом:
finish_dttm, agent_login, comm_finish_dttm,
duratoin_sec AS duration_sec, hit_rk, hit_status_result_id, -- исправление ошибки в ситаксисе "duration"
hid, using_flg, employee_rk, org_group_rk, hit_status_result_desc, queue_id,
queue_desc, org_management_rk, group_nm, management_nm, horoscope
FROM cte1;
```

```python
%%sql -- Результирующая таблица
SELECT *
FROM QWE;
```

# Задание 2
Необходимо посчитать следующие значения:

* Общее количество звонков со статусом «Дозвон, Успешно»;
* Продажная конверсия, по формуле:

  $$ Конверсия = \frac{N_{Дозвон, Успешно}}{N_{Дозвон, Успешно} + N_{Дозвон, Отказ}}$$
* Средняя длительность одной коммуникации, закрытой статусом «Дозвон, Успешно»

Ответы необходимо вписать в соответствующую таблицу в фаиле "Техническое задание"


## Решение

```python
%%sql --Общее количество звонков со статусом «Дозвон, Успешно»
SELECT COUNT (CASE WHEN hit_status_result_id = 3 THEN 1 ELSE NULL END) AS n_success
FROM QWE;
```

```python
%%sql -- Конверсия продаж по формуле
SELECT ROUND(
	(
		(CAST(COUNT (CASE WHEN hit_status_result_id = 3 THEN 1 ELSE NULL END) AS FLOAT) / (
		 CAST(COUNT (CASE WHEN hit_status_result_id = 3 THEN 1 ELSE NULL END) AS FLOAT) +
		 CAST(COUNT (CASE WHEN hit_status_result_id = 1 THEN 1 ELSE NULL END) AS FLOAT)   )
	  )*100
	)
  			, 2) AS conversion_r_percent
FROM QWE;
```

```python
%%sql -- Средняя длительность одной коммуникации  при успешном дозвоне
WITH cte4 AS (
  SELECT *
  FROM QWE
  WHERE hit_status_result_id = 3
  ) 																					-- Оконная функция обсчитывается раньше фильтрации (where), поэтому необходим CTE
SELECT ROUND(
  avg(CAST(duration_sec AS FLOAT))
 				 , 2) AS avg_com_time_success
FROM cte4;
```

**Пример Аномалии**

```python
%%sql -- АНОМАЛИЯ в работе колл-центра
SELECT DISTINCT  agent_login,hit_status_result_desc,
	COUNT(*) OVER(PARTITION BY agent_login, hit_status_result_desc) AS cert_res_status, COUNT(*) OVER(PARTITION BY agent_login) AS N_all, COUNT(*) OVER(PARTITION BY agent_login, hit_status_result_desc) / COUNT(*) OVER(PARTITION BY agent_login)*100 AS percent
FROM QWE
ORDER BY percent DESC, cert_res_status DESC
LIMIT 5;
```
